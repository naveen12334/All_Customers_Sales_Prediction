```python
import pandas as pd
import numpy as np
df=pd.read_csv("before_prediction_sale.csv")
df.head()
```

```python
df.drop('Account Name',axis=1,inplace=True)
df
```

```python
df.fillna(0, inplace = True)
df.head(40)
```

```python
df[['Account Number', 'Apr MTD Sales', 'Aug MTD Sales', 'Feb MTD Sales',
       'Jan MTD Sales', 'July MTD Sales', 'June MTD Sales', 'Mar MTD Sales',
       'May MTD Sales', 'Sept MTD Sales']]=df[['Account Number', 'Apr MTD Sales', 'Aug MTD Sales', 'Feb MTD Sales',
       'Jan MTD Sales', 'July MTD Sales', 'June MTD Sales', 'Mar MTD Sales',
       'May MTD Sales', 'Sept MTD Sales']].astype(int)
df.head(40)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df.columns
```

```python
X_train = df[['Jan MTD Sales','Feb MTD Sales','Mar MTD Sales','Apr MTD Sales','May MTD Sales','June MTD Sales','July MTD Sales']]
Y_train = df["Aug MTD Sales"]

X_test = df[['Feb MTD Sales','Mar MTD Sales','Apr MTD Sales','May MTD Sales','June MTD Sales','July MTD Sales','Aug MTD Sales']]
Y_test = df["Sept MTD Sales"]
```

```python
df.dtypes
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

DT_Classifier = DecisionTreeClassifier()
DT_Classifier = DT_Classifier.fit(X_train, Y_train)

Y_train_pred = DT_Classifier.predict(X_train)
Y_test_pred = DT_Classifier.predict(X_test)


print(f"Training Accuracy Score : {accuracy_score(Y_train, Y_train_pred)}")
print(f'Testing Accuracy Score: {accuracy_score(Y_test, Y_test_pred)}')
```

```python
X = df[['Mar MTD Sales','Apr MTD Sales','May MTD Sales','June MTD Sales','July MTD Sales','Aug MTD Sales','Sept MTD Sales']]
df["Oct_predicted_sale"] = DT_Classifier.predict(X)
df.head(40)
```

```python
df.tail(40)
```

```python
cols = df.columns
for i in range(len(cols)-1):
  if cols[i]!='Account Number' and cols[i]!=cols[i+1]:
    plt.xlabel(cols[i])
    plt.ylabel(cols[i+1])
    plt.title(f"{cols[i]}  Vs  {cols[i+1]}")
    plt.scatter(df[cols[i]], df[cols[i+1]])
    plt.show()
```

```python
df.to_csv('oct_predicted_sale.csv',index=False)
```
